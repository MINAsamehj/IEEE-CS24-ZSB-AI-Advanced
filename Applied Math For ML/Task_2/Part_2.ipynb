## using `numpy`

```python
norm = np.random.normal(0,scale = 1, size= (1000) )

plt.hist(norm)
plt.grid()
plt.show()

print(
    f"""
sample mean =  {np.mean(norm)}
sample vriance = {np.var(norm)}
    
    """
)
```

## using `scipy`

```python
random = stats.norm.rvs(loc=0, scale=1, size=1000, random_state=None)
plt.hist(random , density=True)
plt.grid()
x = np.linspace(-4, 4,1000)

pdf = stats.norm.pdf(x, loc=0, scale=1) 
plt.plot( x, pdf , color='red')
```

- Compute the true "analytical" PDF using [scipy.stats.norm.pdf](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html). Plot the true PDF on the top of the histogram figure.
- Compute the CDF using [scipy.stats.norm.cdf](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html), then plot it.

```python
random = stats.norm.rvs(loc=0, scale=1, size=1000, random_state=None)
plt.hist(random , density=True)
plt.grid()
x = np.linspace(-4, 4,1000)

pdf = stats.norm.pdf(x, loc=0, scale=1) 
plt.plot( x, pdf , color='red')


cdf = stats.norm.cdf(x, loc=0 , scale=1)
plt.plot(x, cdf , color = 'orange')
```
